```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame(np.random.randint(0,25,size=(10, 4)), columns=['Column A', 'Column B', 'Column C', 'Column D'])
```

```python
df
```

```python
df_class = pd.DataFrame(np.random.randint(0,1,size=(10, 1)), columns=['Output Column'])
```

```python
df = df.join(df_class)
df
```

```python
X = df.drop('Output Column', axis=1)
y = df['Output Column']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=42, shuffle=False)
```
