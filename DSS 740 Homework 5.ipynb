```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import mglearn as mglearn
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, ward
```

# Question 4

```python
X, y = make_blobs(random_state=0, n_samples=500)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
print("Cluster membership:\n{}".format(kmeans.labels_))
```

```python
print("k-Means Prediction membership:\n{}".format(kmeans.predict(X)))
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
assignment = agg.fit_predict(X)
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], assignment)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
mglearn.plots.plot_agglomerative()
```

```python
dbscan = DBSCAN()
```

```python
clusters = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap=mglearn.cm2, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

# Question 5

```python
wine = datasets.load_wine()
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['Target'] = wine.target
df.head()
```

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

```python
ss = StandardScaler()
ss.fit_transform(X)
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(X)
```

```python
X_pca = pca.transform(X)
print("Original shape: {}".format(str(X.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))
```

```python
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], wine.target)
plt.legend(wine.target_names, loc="best")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

# Question 6

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Target'] = iris.target
df.head()
```

```python
X = df.drop('Target', axis=1)
y = df['Target']
```

```python
ss = StandardScaler()
ss.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
print("Cluster membership:\n{}".format(kmeans.labels_))
```

```python
print("k-Means Prediction membership:\n{}".format(kmeans.predict(X)))
```

```python
mglearn.discrete_scatter(X.iloc[:, 0], X.iloc[:, 1], kmeans.labels_, markers='o')
```

```python
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical.labels_ = hierarchical.fit_predict(X)
```

```python
linkage_array = ward(X)
dendrogram(linkage_array)
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [7.25, 7.25], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')
plt.xlabel("Index")
plt.ylabel("Cluster distance")
```

```python
agg = AgglomerativeClustering(n_clusters=3)
```

```python
assignment = agg.fit_predict(X)
```

```python
mglearn.discrete_scatter(X.iloc[:, 0], X.iloc[:, 1], assignment)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
dbscan = DBSCAN()
```

```python
clusters = dbscan.fit_predict(X)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=clusters, cmap=mglearn.cm2, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
dbscan.labels_ = dbscan.fit_predict(X)
```

```python
pca = PCA(n_components=3)
```

```python
X_pca = pca.fit_transform(X)
kmeans_pca = KMeans(n_clusters=3)
```

```python
pca_labels = kmeans.fit_predict(X_pca)
```

```python
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], kmeans.labels_, markers='o')
```

```python
silhouette_kmeans = silhouette_score(X, kmeans.labels_)
silhouette_hierarchical = silhouette_score(X, hierarchical.labels_)
silhouette_dbscan = silhouette_score(X, dbscan.labels_)
silhouette_agg = silhouette_score(X, agg.labels_)
silhouette_pca = silhouette_score(X_pca, kmeans.labels_)
print("silhouette_kmeans: ", silhouette_kmeans)
print("silhouette_hierarchical: ", silhouette_hierarchical)
print("silhouette_agg: ", silhouette_agg)
print("silhouette_dbscan: ", silhouette_dbscan)
print("silhouette_PCA: ", silhouette_pca)
```
